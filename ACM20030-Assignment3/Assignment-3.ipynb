## Question 1: Linear least-square fitting
This questions asks you to write a function to find the best linear fit to a dataset and then apply that function to a few data sets. As always when testing a new piece of code, you should first test the code on a small data set for which you know the result. The normal equations for linear least-squares fitting are given by:  
  
$$
  \begin{align*}
    m \sum x_i^2 + c \sum x_i &= \sum x_i y_i \\
    m \sum x_i + cn &= \sum y_i
  \end{align*}
$$
  
Write these simultaneous equations in matrix for $A.x = b$, where $A$ is a $2\times2$ matrix and $x=\{m,c\}$ and $b$ are column vectors. Note that for a $2\times2$ matrix 
$$
A = \begin{pmatrix}
    a  & b \\
    c  & d 
\end{pmatrix}
$$
the inverse is given by
$$
A^{-1} = \frac{1}{ad-bc}\begin{pmatrix}
    d  & -b \\
    -c  & a 
\end{pmatrix}
$$

### Q1a.

Write a function that takes in two arrays of data for the x and y values and returns the values of $m$ and $c$ from for the linear least squares fit. Call the function `LinearLeastSquares`. You might want to use the function `np.dot(A,b)` which will take the dot product of a matrix, $A$ and a vector, $b$.

```python
def LinearLeastSquares(data):
    
    #Defining the x and y arrays for the np.sum() function, and defining n.
    x=np.array(data[:,:1])
    y=np.array(data[:,1:])
    n=len(data)
    
    #Defining the coefficient matrix for m and c, the matrix B, and the inverse of the matrix A. Ensuring in all cases that
    #the arrays keep their required shape
    A = np.array([np.sum(arr) for arr in [x**2, x, x, n]]).reshape((2,2))
    B = np.array([np.sum(arr) for arr in [x*y , y]]).reshape((2,1))
    Ainv=(1/(A[0][0]*A[1][1]-A[0][1]*A[1][0]))*np.matrix([[A[1][1],-1*A[0][1]],[-1*A[1][0],A[0][0]]])
    
    #Defining the Resulting Matrix which contains the values for m and c
    ResMat=np.dot(Ainv,B)
    
    #Printing the values for m and c explicitly, and assinging them to variables so they can be placed in a tuple when
    #the function is executed
    m=float(ResMat[0])
    c=float(ResMat[1])
    print("m is equal to ",m," and c is equal to ",c)
    return m, c
```

## Question 2: second-order ODEs

Use the below second-order version of the (fixed-step) RK2 function for this question.

```python
def RK2(f, dx, x0, y0, imax):
    output = np.empty((imax, 3))
    i = 0
    xi = x0
    yi = y0
    while(i < imax):
        k1 = dx*f(xi, yi)
        k2 = dx*f(xi + dx, yi + k1)
        yi = yi + 0.5*(k1 + k2)
        xi += dx
        output[i, 0] = xi
        output[i, 1] = yi[0]
        output[i, 2] = yi[1]
        i += 1
    return output
```

You should find the systems with initial values $u(0) = \{0,2\}$ do not evolve as the right-hand side equals zero in these cases. These two values are called *equilibrium points*.
```python
def BackEulerMethod(df, yreal, dx, x0, y0, imax, printSteps=False, plotSteps=True, uColor='red'):
    i = 0
    xi = x0
    yi = y0
    
    steps = np.zeros((imax+1,2))
    steps[0,0] = x0
    steps[0,1] = y0
    
    while i<imax:
        yi = (yi+dx*xi)/(1+dx)
        xi += dx
        i += 1
        
        steps[i, 0] = xi
        steps[i, 1] = yi
    
    if(plotSteps)==True:
        plt.scatter(steps[0:,0], steps[0:,1],color=uColor,linewidth=5,label="Numerical Solution")
        
    return [xi,yi]

def yreal(x):
    return -1+2*np.e**(-x)+x

def df(x,y):
    return -y+x

resBEM=BackEulerMethod(df, yreal, 0.1, 0, 1, 20)
print(resBEM)
```
